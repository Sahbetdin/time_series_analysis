### <a id='stationarity'>Stationarity</a>

В визуальном анализе мы уже провели первичный [тест](#viz1) на стационарность. выполним тест Дики-Фуллера для проверки стационарности.

```python
from statsmodels.tsa.stattools import adfuller
def test_statonarity(col):
    print("results of Dickey-Fuller Test:")
    dftest = adfuller(col, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index = ['Test Statistic','p-value','#Lags Used', 'Number of Observations Used'])
    for key, value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    print(dfoutput)
```

***Прим.*** Следующая ячейка выполняется долго.

```python
test_statonarity(df.y1)
```

Так как значение p-value ~0.47, мы можем опровергнуть нуль гипотезу о том, что данные стационарны.

Как известно, нестационарную функцию можно разложить на несолько составляющих, которые включают в себя трендовую, сезонную и рандомную часть, которая может иметь свои причины:

## C. Обучение линейной модели

### <a id='mat_form'>Математическая формулировка задачи линейной регрессии</a>

Давайте возьмем данные Matr00, которые есть наши данные при $s_1 = 0, s_2 = 0$ и применим к ним линейную модель. Делим Matr00 на части по 4/5 в обучающей и 1/5 в тестовой выборках. Потом то же самое применим к данным с другим значениями $s_1$, $s_2$, то есть к другим матрицам Matr.

```python
import sklearn.model_selection as model_selection
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
import math
```

```python
X = Matr00.drop(['x2','x7','y1'],axis=1)
y = Matr00.drop(['x1','x2','x3','x4','x5','x6','x7','x8'],axis=1)
```

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.2, random_state=17)
n_train = len(X_train)
n_test = len(X_test)
print('There are', n_train, 'elements in training set.')
print('There are', n_test, 'elements in testing set.')
```

```python
model = LinearRegression()
model.fit(X_train, y_train);
```

Посмотрим на начальное значение $y_1$ и коэффициенты при $x_1...x_8$:

```python
print(X_train.columns)
print(model.intercept_, model.coef_)
```

Так как значения $x_1...x_8$ находятся примерно в одном коридоре, то полученные выше коэффициенты в принципе отражают действительность о вкладе каждого из переменных в $y_1$.
Ниже представлены соответсвующие макс и мин значения

```python
print('min x1 =', min(X_train.x1), ', max x1 =', max(X_train.x1))
print('min x3 =', min(X_train.x3), ', max x3 =', max(X_train.x3))
print('min x4 =', min(X_train.x4), ', max x4 =', max(X_train.x4))
print('min x5 =', min(X_train.x5), ', max x5 =', max(X_train.x5))
print('min x6 =', min(X_train.x6), ', max x6 =', max(X_train.x6))
print('min x8 =', min(X_train.x8), ', max x8 =', max(X_train.x8))
```

**Как преварительный итог:** видим, что $y_1$ сильно зависит от $x_6$ и далее от $x_3$.

Посмотрим, как полученная модель предсказывает значения для тестовой выборки:

```python
y_pred = model.predict(X_test)
```

Строить график данных от времени бессмысленно, так как данные выбраны случайным образом.

```python
r_sq = model.score(X, y)
mse = mean_squared_error(y_test.values, y_pred)
print('coefficient of determination R^2 =', r_sq)
print('MSE =', mse)
print('RMSE =', math.sqrt(mse))
print('MAE =', mean_absolute_error(y_test.values, y_pred))
```

Самый главный вывод, который мы можем сделать, что ~76% зависимостей может быть объяснено полученной линейной моделью.

Ниже построена <a id='hist'>гистограмма</a> остатков, а также нормальное распределение, которое аппроксимирует гистограмму.

```python
from scipy.stats import norm

# Fit a normal distribution to the data:
mu, std = norm.fit(dy.values);

# Plot the histogram.
plt.hist(dy.values, bins=20, density=True, alpha=0.6, color='blue');

# Plot the PDF.
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std)
plt.plot(x, p, 'k', linewidth=2)
title = "Fit results: mu = %.2f,  std = %.2f" % (mu, std)
plt.title(title)

plt.show()
```

```python
import statsmodels.api as sm
import scipy.stats as stats
```

```python
ddy = np.reshape(dy.values, (-1, ))
fig = sm.qqplot(ddy, stats.t, fit=True, line="45", scale=1)
plt.show()
```
